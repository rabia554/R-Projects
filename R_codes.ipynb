```r
hist(lynx,
 breaks = 14,        #suggests 14 bins(14 bars)
 col = "red",        #Color for Histogram
 freq = FALSE,          #Axis shows density not frequency
 main = paste("Histogram of Annual Canadian Lynx","Trappings, 1821-1934"),
 xlab = "Number of Linux Trapped")
 curve(dnorm(x,mean = mean(lynx),sd = sd(lynx)),
 col="thistle",   #Color of curve
 lwd=2,           #Line width of 2 pixels
 add = TRUE)       #Superimpose on other graph

 #Add two kernel density estimators
 lines (density(lynx), col="blue",lwd=2)
 lines(density(lynx, adjust = 3),col = "purple", lwd = 2)
 #Add a rug plot
 rug(lynx, lwd = 2,col = "gray")
```
